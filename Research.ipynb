# **Light Curves**

```python
!pip install lightkurve
from lightkurve import search_lightcurvefile

# List of TIC IDs or Star names for each target
targets = [
    "HD 202772 A", "HD 2685", "TOI-150", "TOI-163", "TOI-163",
    "TOI-172", "TOI-677", "HIP 94235", "HIP 67522", "HIP 67522",
    "TIC 172900988"
]

# Downloading the light curve data for each target
for target in targets:
    search_result = search_lightcurvefile(target)
    lc_file = search_result.download()
    lc = lc_file.PDCSAP_FLUX.remove_nans().normalize()

    # Plotting the light curve (optional)
    lc.plot(title=f'Light Curve for {target}')
```
